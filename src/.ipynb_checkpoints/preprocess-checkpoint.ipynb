# 3. Feature Engineering

## 3-1. Embarked

```python
df = pd.get_dummies(df, columns=['Embarked'], prefix='Em')
```

## 3-1. Cabin

```python
df['nCabin'] = pd.Series([1 + x.count(' ') if not pd.isnull(x) else 0
                          for x in df['Cabin']])
df['Cabin'] = pd.Series([x[0] if not pd.isnull(x) else 'X' for x in df['Cabin']])
df.head()
```

```python
g = sns.barplot(x='Cabin', y='Survived', data=df)
```

```python
df = pd.get_dummies(df, columns=['Cabin'], prefix='Cab')
```

## 3-2. Name

```python
df['Name_length'] = df['Name'].str.len()
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
g = sns.factorplot(x='Title', y='Survived', data=df, kind='bar', size=6)
g.set_xticklabels(rotation=45)
g = sns.factorplot(x='Title', data=df, kind='count', size=6)
g.set_xticklabels(rotation=45);
```

```python
df['Title'].replace(['Mlle', 'Ms', 'Mrs', 'Mme'], 'Miss', inplace=True)
df['Title'].replace(['Countess', 'Dona', 'Lady', 'Major', 'Sir', 'Jonkheer'],
                     'Rare', inplace=True)
df['Title'].replace(['Capt', 'Col', 'Don', 'Dr', 'Rev'],
                     'Officer', inplace=True)
df[['Title', 'Survived']].groupby('Title').mean()
```

```python
g = sns.factorplot(x='Title', y='Survived', data=df, kind='bar', size=6)
g = sns.factorplot(x='Title', data=df, kind='count', size=6)
```

```python
df.drop('Name', axis=1, inplace=True)
df = pd.get_dummies(df, columns=['Title'], prefix='Title')
```

```python
df.head()
```

## 3-3. Ticket

```python
df['sameTicket'] = df.groupby('Ticket')['Ticket'].transform('count')

df['Ticket'] = pd.Series([x.replace('.', '').replace('/', '').strip().split()[0]
                          if not x.isdigit() else 'X' 
                          for x in df['Ticket']])
```

```python
df = pd.get_dummies(df, columns=['Ticket'], prefix='T')
df.head()
```

## 3-4. Family

```python
df['Family'] = df['SibSp'] + df['Parch'] + 1

g = sns.countplot(x='Family', hue='Survived', data=df)
```

```python
df.drop('PassengerId', axis=1, inplace=True)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

# 4. Filling missing Data

3. Age

```python
from sklearn.ensemble import RandomForestRegressor
def fill_age(df):
    mask = df['Age'].isnull()
    x = df[mask == False].drop(['Age', 'Survived'], axis=1)
    y = df.loc[mask == False, 'Age']
    miss_age = df[mask].drop(['Age', 'Survived'], axis=1)
    rf = RandomForestRegressor()
    rf.fit(x, y)
    df.loc[mask, 'Age'] = rf.predict(miss_age)

fill_age(df)
```

```python
from utils import encode_as_band
df = encode_as_band(df, ['Age', 'Fare', 'Name_length'], 'qcut')
df.head()
```

```python
df['isChild'] = (df['Age_Band'] == 0).astype(int)
df.loc[df['Family'] == 1, 'Family'] = 0
df.loc[(df['Family'] > 1) & (df['Family'] <= 4), 'Family'] = 1
df.loc[df['Family'] > 4, 'Family'] = 2
df.loc[df['sameTicket'] == 1, 'sameTicket'] = 0
df.loc[(df['sameTicket'] > 1) & (df['sameTicket'] <= 4), 'sameTicket'] = 1
df.loc[df['sameTicket'] > 4, 'sameTicket'] = 2
df.head()
```

# Onehot Encoding

```python
from utils import onehot_encoding
df = onehot_encoding(df, ['Pclass', 'Age_Band', 'Fare_Band', 'Name_length_Band'])
df.head()
```
